```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 定义 H(z) 函数
def H(z, H0, m):
    return H0 / np.sqrt(2) * np.sqrt(1 + (1 + z)**(2 * m))
```

```python
z = np.linspace(-1, 4, 200)  # 红移范围
plt.plot(z,H(z,66,1.28))
```

```python
# 设置参数
H0 = 66  # km/s/Mpc
alpha = 1
n = 1
m = 1.28
c=1
z = np.linspace(-1, 4, 200)  # 红移范围

H_z = H(z, H0, m)
dot_H = np.gradient(H_z,z)*(-H_z*(1+z))
ddot_H=np.gradient(dot_H,z)*(-H_z * (1 + z))
plt.plot(z,dot_H)
plt.show()

plt.plot(z,ddot_H)
plt.show()
```

```python
rde=(3*H_z**2)
pde=alpha*(-6*H_z**2)**n*(n-1/2)+(alpha*n)/3*dot_H*(-6)**n*H_z**(2*n-2)*(2*n-1)
wde=pde/rde
plt.plot(z,pde)
```

```python


# 设置参数
H0 = 66  # km/s/Mpc
alpha = 1
n = 1
m = 1.28
c=1
z = np.linspace(-1, 4, 200)  # 红移范围

H_z = H(z, H0, m)


#dot_H=H0*m*(1+z)**(-1+2*m)/(np.sqrt(2)*np.sqrt(1+(1+z)**(2*m)))*(-H_z*(1+z))
dot_H = -np.gradient(H_z,z)*H_z*(1+z)
ddot_H=np.gradient(dot_H,z)*(-H_z * (1 + z))
# dot_H = -(H0**2)/2 *m * (1+z)**(2*m)

rde=(3*H_z**2)
pde=alpha*(-6*H_z**2)**n*(n-1/2)+(alpha*n)/3*dot_H*(-6)**n*H_z**(2*n-2)*(2*n-1)
wde=pde/rde
# 绘制图像
plt.figure(figsize=(8, 6))
plt.plot(z, wde)
plt.xlabel("Redshift (z)")
plt.ylabel("w(z)")
plt.title("Equation of State for Dark Energy")
plt.grid()  # 添加网格
# 添加 z=0 的垂直线
plt.axvline(x=0, color='r', linestyle='--', label='z=0')
plt.legend()
# 在 z=0 处添加 w(z) 的值
z_zero_idx = np.argmin(np.abs(z))
w_z0 = wde[z_zero_idx]
plt.text(0, w_z0, f"w(z=0) = {w_z0:.4f}", ha='center', va='bottom', color='r')
# 在 z=0 处添加标记点
plt.plot(0, w_z0, 'ro', markersize=10)
plt.show()

plt.plot(z, pde)
plt.show()

plt.plot(z,rde)
plt.show()

cs2=alpha*(-6)**n*2*n*(n-1/2)*H_z**(2*n-2)+alpha*n/3*(-6)**n*(2*n-1)*H_z**(2*n-3)*ddot_H/dot_H+alpha*n/3*(-6)**n*(2*n-1)*(2*n-2)*dot_H*H_z**(2*n-4)
plt.plot(z,-cs2)
plt.show()
```
