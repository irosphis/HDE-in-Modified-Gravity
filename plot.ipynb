$$
\begin{align*}
\rho &=\frac{f}{2}-6f_Q H^2-\frac{2f_T}{1+f_T}(\dot{f}_QH+f_Q \dot{H}) \\
p &=-\frac{f}{2}+6f_Q H^2+2(\dot{f}_QH+f_Q \dot{H})
\end{align*}
$$
其中$f=\alpha Q+\beta T$，$Q=6H^2$，$T=-\rho+3p$，$w=\frac{p}{\rho}$，$H=k(1+(1+z)^\xi)$

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数
alpha = 1.0
beta = 1.0
k = 0.1
xi = 2.4

# 定义红移范围
z = np.linspace(-1, 2, 100)

# 计算 H 和 Q
H = k * (1 + (1 + z)**xi)  # H(z) 的定义
Q = 6 * H**2  # 定义 Q
T = -rho-3*p

# 计算 f 和 f_Q
f = alpha * Q + beta * T  # 这里可以根据具体情况修改 T 的计算方式
f_Q = alpha  # 假设 f_Q = α，具体形式可能需要根据 T 的定义来调整

# 计算 H 的导数
dot_H = np.gradient(H, z)  # 计算 H 的导数

# 计算 T
T = - (alpha * Q + beta * (1 - Q)) + 3 * (dot_H + f_Q * dot_H)

# 计算 rho 和 p
rho = (f / 2) - 6 * f_Q * H**2 - (2 * f_Q / (1 + f_Q)) * (dot_H * H + f_Q * dot_H)
p = - (f / 2) + 6 * f_Q * H**2 + 2 * (dot_H * H + f_Q * dot_H)

# 计算 w
w = p / rho

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(z, w, label='w(z)')
plt.title('Relationship between w and Redshift z')
plt.xlabel('Redshift z')
plt.ylabel('w')
plt.axhline(0, color='gray', linestyle='--')  # 添加 w = 0 的参考线
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 定义常量
c = 1
m = 1.0
n = 1.1
alpha = 0.1
wde = -0.7

# 定义右侧的微分方程
def differential_eq(z, y):
    H, dH = y  # y[0] = H, y[1] = dH/dz
    term1 = 6 ** (-1 + n) * m * (-1 + 2 * n) * (H**(-2 + 2 * n))
    term2 = 3 * (1 + alpha) * H**2
    term3 = -n * (1 + z) * (2 + alpha) * H * dH
    numerator = term1 * (term2 + term3)
    denominator = wde * (-1 + alpha + 2 * alpha**2)
    rhs = numerator / denominator - 3 * c**2 * H**2
    return [dH, rhs / (n * (1 + z) * (2 + alpha))]  # 返回 [H', dH']

# 初值和范围
z_start = 0
z_end = 2
H0 = 70.0
dH0 = -5.0  # 初始导数（需要估计或假设）

# 求解
solution = solve_ivp(differential_eq, [z_start, z_end], [H0, dH0], dense_output=True)

# 提取解
z_vals = np.linspace(z_start, z_end, 500)
H_vals = solution.sol(z_vals)[0]  # 第一个分量是 H(z)

# 可视化
plt.plot(z_vals, H_vals, label="H(z)")
plt.xlabel("z")
plt.ylabel("H(z)")
plt.title("Numerical Solution for H(z)")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.optimize import minimize
from scipy.integrate import solve_ivp

# 定义观测数据
z_obs = np.array([0.1, 0.2, 0.3, 0.5, 0.6, 0.7, 0.8, 1.0, 1.2, 1.5, 1.8, 2.0])  # 扩展红移值
H_obs = np.array([69.0, 72.2, 83.1, 107.5, 125.0, 133.5, 140.0, 150.0, 160.0, 170.0, 180.0, 200.0])  # 对应的H(z)
sigma_H = np.array([1.2, 1.0, 1.5, 2.0, 2.2, 2.0, 2.5, 2.8, 3.0, 3.2, 3.5, 4.0])  # 对应的误差


# 模型函数
def H_model(params, z_vals):
    n, alpha, m, wde, c = params
    # 定义微分方程与初值，类似之前代码
    def differential_eq(z, y):
        H, dH = y
        term1 = 6 ** (-1 + n) * m * (-1 + 2 * n) * (H**(-2 + 2 * n))
        term2 = 3 * (1 + alpha) * H**2
        term3 = -n * (1 + z) * (2 + alpha) * H * dH
        numerator = term1 * (term2 + term3)
        denominator = wde * (-1 + alpha + 2 * alpha**2)
        rhs = numerator / denominator - 3 * c**2 * H**2
        return [dH, rhs / (n * (1 + z) * (2 + alpha))]
    
    # 解方程
    solution = solve_ivp(differential_eq, [z_vals[0], z_vals[-1]], [70.0, -5.0], t_eval=z_vals)
    return solution.y[0]  # 返回 H(z) 数值

# 定义误差函数
def chi_squared(params):
    H_pred = H_model(params, z_obs)
    chi2 = np.sum(((H_obs - H_pred) / sigma_H) ** 2)
    return chi2

# 初始参数猜测
initial_guess = [1.1, 0.1, 1.0, -1.0, 1.0]

# 优化参数
result = minimize(chi_squared, initial_guess, method='Nelder-Mead')
best_fit_params = result.x

print("Best fit parameters:", best_fit_params)
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def compute_H(z_vals, H0, c, n, m, alpha, wde):
    """
    计算 H(z) 的数值解
    参数：
        z_vals : array-like
            要计算的 z 值
        H0 : float
            z=0 时的哈勃参数初值 H(0)
        c : float
            参数 c
        n : float
            参数 n
        m : float
            参数 m
        alpha : float
            参数 alpha
        wde : float
            参数 wde
    返回：
        H_vals : array-like
            对应于 z_vals 的 H(z) 数值解
    """
    # 定义常微分方程组
    def ode_system(z, y):
        H, H_prime = y
        # 改写原方程得到 H''(z) 的形式
        H_double_prime = (3 * c**2 * H - (6**(-1 + n) * m * (-1 + 2 * n) * (H**2)**(-1 + n) *
                           (3 * (1 + alpha) * H**2 - n * (1 + z) * (2 + 3 * alpha) * H_prime)) /
                          (wde * (1 + 3 * alpha + 2 * alpha**2))) / H
        return [H_prime, H_double_prime]

    # 初始条件和积分范围
    z_span = (z_vals[0], z_vals[-1])  # z 的积分范围
    y0 = [H0, 0]  # 初始值：H(z=0) = H0，H'(z=0) = 0

    # 使用 solve_ivp 进行积分
    solution = solve_ivp(ode_system, z_span, y0, t_eval=z_vals)

    # 返回数值解
    return solution.y[0]

# 示例：使用该函数计算 H(z)
if __name__ == "__main__":
    # 输入参数
    z_vals = np.linspace(0, 2, 100)  # z 的取值范围
    H0 = 70  # 初始哈勃参数 H(z=0)
    c = 1.0
    n = 0.5
    m = 1.0
    alpha = 0.1
    wde = 0.7

    # 计算 H(z)
    H_vals = compute_H(z_vals, H0, c, n, m, alpha, wde)

    # 绘制结果
    plt.plot(z_vals, H_vals, label='H(z)')
    plt.xlabel('z')
    plt.ylabel('H(z)')
    plt.title('Numerical Solution of H(z)')
    plt.legend()
    plt.grid()
    plt.show()
```
